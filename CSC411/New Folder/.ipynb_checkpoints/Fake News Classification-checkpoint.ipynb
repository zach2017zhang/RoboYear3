```python
def split_dataset(dataset, label=1, train_ratio=0.7, valid_ratio=0.15, test_ratio=0.15, shuffle=True):
    word_lists = [data.split() for data in dataset]
    word_lists_labels = [(w_list, label) for w_list in word_lists]
    
    idx = [i for i in range(len(word_lists))]
    np.random.seed(0)
    if shuffle:
        np.random.shuffle(idx)
    shuffled_word_lists = [word_lists_labels[i] for i in idx]
    
    num = len(shuffled_word_lists)
    train = shuffled_word_lists[:int(num * train_ratio)]
    valid = shuffled_word_lists[int(num * train_ratio) : -int(num * test_ratio)]
    test = shuffled_word_lists[-int(num * test_ratio): ]
    return train, valid, test
```

```python
class NaiveBayesClassifier(object):
    """ Naive Bayes classifier for fake and real news 
    """
    def __init__(self, datasets, r_vocab, f_vocab, param_path="NBparams.pkl", reuse=False):
        # smoothing priors 
        self.m = 100
        self.p = 0.001
        
        # datasets
        self.r_vocab = list(set(r_vocab))
        self.f_vocab = list(set(f_vocab))
        
        self.r_train = datasets["r_train"]
        self.r_valid = datasets["r_valid"]
        self.r_test = datasets["r_test"]
        self.f_train = datasets["f_train"]
        self.f_valid = datasets["f_valid"]
        self.f_test = datasets["f_test"]
        
        self.count_word_freq()
        self.datasets = {
            "train": self.r_train + self.f_train, 
            "valid": self.r_valid + self.f_valid,
            "test": self.r_test + self.f_test
        }
        
        # count params 
        if os.path.exists(param_path) and reuse:
            self.load_params(param_path)
        else:
            self.param_count()
            self.save_params(path=param_path)
        
        
    def count_word_freq(self):
        self.r_freq_dict = {}
        self.f_freq_dict = {}
        for sample, label in self.r_train:
            for w in list(set(sample)):
                self.r_freq_dict[w] = 1 if w not in self.r_freq_dict else self.r_freq_dict[w] + 1
        for sample, label in self.f_train:
            for w in list(set(sample)):
                self.f_freq_dict[w] = 1 if w not in self.f_freq_dict else self.f_freq_dict[w] + 1

        
    def param_count(self):
        # class priors 
        total = len(self.r_train) + len(self.f_train)
        self.r_prior = len(self.r_train) / total 
        self.f_prior = len(self.f_train) / total 
        
        # condition word prob
        self.word_cond_prob = {1: {}, 0: {}}
        self.smoothing_token = {
            1: self.m * self.p / (len(self.r_train) + self.m),
            0: self.m * self.p / (len(self.f_train) + self.m),
        }
        for w in self.r_freq_dict:
            self.word_cond_prob[1][w] = (self.r_freq_dict[w] + self.m * self.p) / (len(self.r_train) + self.m)
        for w in self.f_freq_dict:
            self.word_cond_prob[0][w] = (self.f_freq_dict[w] + self.m * self.p) / (len(self.f_train) + self.m)
            
    
    def save_params(self, path="NBparams.pkl"):
        params = {
            "r_prior": self.r_prior,
            "f_prior": self.f_prior,
            "word_cond_prob": self.word_cond_prob
        }
        with open(path, "wb") as f:
            pickle.dump(params, f)
            
            
    def load_params(self, path="NBparams.pkl"):
        with open(path, "rb") as f:
            params = pickle.load(f)
        self.r_prior = params["r_prior"]
        self.f_prior = params["f_prior"]
        self.word_cond_prob = params["word_cond_prob"]
            
        
    def set_hyperparams(self, hyperparamset, display=True):
        self.m, self.p = hyperparamset[0], hyperparamset[1]
        if display:
            print("current m: ", self.m, end=" ")
            print("current p: ", self.p)
        
        
    def inference_single_step(self, x, smoothing=True, err=10**(-10)):
        """ infer news label
        """
        # smoothing token 
        if smoothing:
            self.smoothing_token = {
                1: self.m * self.p / (len(self.r_train) + self.m),
                0: self.m * self.p / (len(self.f_train) + self.m),
            }
        
        # numerator of real news class 
        #r_prob_list = np.asarray([self.word_cond_prob[1][key] if key in x else 1.0-self.word_cond_prob[1][key] for key in self.word_cond_prob[1]])
        r_prob_list = np.asarray([self.word_cond_prob[1][w] if w in self.word_cond_prob[1] else self.smoothing_token[1] for w in x])
        r_prob_list_absence = np.asarray([1.0-self.word_cond_prob[1][key] for key in self.word_cond_prob[1] if key not in x])
        real_prob =  np.exp(np.sum(np.log(r_prob_list))) * np.exp(np.sum(np.log(r_prob_list_absence))) * self.r_prior
        #print(real_prob)

        # numerator of fake news class 
        #f_prob_list = np.asarray([self.word_cond_prob[0][key] if key in x else 1.0-self.word_cond_prob[0][key] for key in self.word_cond_prob[0]])
        f_prob_list = np.asarray([self.word_cond_prob[0][w] if w in self.word_cond_prob[0] else self.smoothing_token[0] for w in x])
        f_prob_list_absence = np.asarray([1.0-self.word_cond_prob[0][key] for key in self.word_cond_prob[0] if key not in x])
        fake_prob =  np.exp(np.sum(np.log(f_prob_list))) * np.exp(np.sum(np.log(f_prob_list_absence)))* self.f_prior
        #print(fake_prob)
        #print()

        # normalize
        real_prob /= (real_prob + fake_prob)
        #print(real_prob)
        return real_prob
    
    
    def inference_batch(self, dataset):
        """ inference on a batch 
        """
        preds = np.asarray([self.inference_single_step(x) for x, label in dataset])
        return preds 
        
    
    def classify(self, output):
        return np.asarray(output > 0.5).astype(int)
    
    
    def accuracy(self, preds, labels):
        return np.sum(np.asarray(np.equal(self.classify(preds), labels).astype(int))) / float(len(labels))
    
    
    def evaluate(self, mode="train", custom_dataset=None, display=True):
        """ evaluate on a dataset -> list of (word_list, label) pairs
        """
        if mode == "train" or mode == "valid" or mode == "test":
            dataset = self.datasets[mode]
        else:
            dataset = custom_dataset
        preds = self.inference_batch(dataset)
        labels = np.asarray([label for w, label in dataset])
        acc = self.accuracy(preds, labels)
        if display:
            print(mode, " acc ", acc)
        return acc 
    
    
    def tune_hyperparams(self, display=True, save=True):
        """ tune hyperparameters m, p, by default takes best hyperparam set with highest validaion accuacy
        """
        # test on current selction 
        old_m = self.m
        old_p = self.p
        valid_set = self.r_valid + self.f_valid
        acc = self.evaluate(mode="valid", display=False)
        res = {(self.m, self.p): acc}
        
        # grid search 
        cand_m = [100*i for i in range(1, 11)]
        cand_p = [10**(-i) for i in range(7)]
        for m in cand_m:
            for p in cand_p:
                self.m = m
                self.p = p
                acc = self.evaluate(mode="valid", display=False)
                res[(m, p)] = acc
        self.m = old_m
        self.p = old_p
        
        # get best candidate 
        ordered_hyperparamset = sorted(list(res.items()), key=lambda x: x[1], reverse=True)
        choice, acc = ordered_hyperparamset[0][0], ordered_hyperparamset[0][1]
        self.set_hyperparams(choice, display=False)
        if display:
            print("selected m ", choice[0], " p ", choice[1], " acc ", acc)
        if save:
            self.save_params()
        return ordered_hyperparamset
```

```python
word_cond_prob = nb.word_cond_prob
label_priors = [nb.f_prior, nb.r_prior]
smoothings = nb.smoothing_token

# print(label_priors)
# print(smoothings)
```

```python
def label_prob_given_word(word_cond_prob, label_priors, smoothings, word, label):
    # p(label|word) = p(word|label)p(label) / sum (p(word|label)p(label))
    prob1 = word_cond_prob[label].get(word, smoothings[label]) * label_priors[label]
#     print(prob1)
    temp = 0 if label == 1 else 1
    prob2 = word_cond_prob[temp].get(word, smoothings[temp]) * label_priors[temp]
#     print(prob2)
    return prob1 / (prob1 + prob2)
    
def get_influential_words_given_word(word_cond_prob, label_priors, smoothings, label, num):
    bayes_prob_list = [(w, label_prob_given_word(word_cond_prob, label_priors, smoothings, w, label)) for w in list(word_cond_prob[label].keys())]
    ordered_list = sorted(bayes_prob_list, key=lambda x: x[1], reverse=True)
    return {
        "presence": ordered_list[:num],
        "absence": ordered_list[-num:]
    }
```

```python
class LogisticRegressionClassifier(object):
    """ logistic regression for classifying real and fake news 
    """
    def __init__(self, datasets, in_dim=5832, out_dim=1):
        # model hyperparameters
        self.in_dim = in_dim + 1
        self.out_dim = out_dim
        self.reg_strength = 0.0001  # scale of 0.001 is ok with normalized loss 
        
        # parameter initialization
        self.weights = np.zeros(self.in_dim)
        np.random.seed(0)
        # Xavier initialization 
        self.weights[:-1] = np.random.normal(0, 2.0/(self.in_dim-1+self.out_dim), self.in_dim-1)
        self.weights[-1] = np.ones(self.out_dim)
        
        # default data
        self.X_train = self.augment_x(np.concatenate((datasets["r_train_vec_sep"][0], datasets["f_train_vec_sep"][0]), axis=0)) 
        self.t_train = np.concatenate((datasets["r_train_vec_sep"][1], datasets["f_train_vec_sep"][1]), axis=0)
        self.X_valid = self.augment_x(np.concatenate((datasets["r_valid_vec_sep"][0], datasets["f_valid_vec_sep"][0]), axis=0)) 
        self.t_valid = np.concatenate((datasets["r_valid_vec_sep"][1], datasets["f_valid_vec_sep"][1]), axis=0)
        self.X_test = self.augment_x(np.concatenate((datasets["r_test_vec_sep"][0], datasets["f_test_vec_sep"][0]), axis=0)) 
        self.t_test = np.concatenate((datasets["r_test_vec_sep"][1], datasets["f_test_vec_sep"][1]), axis=0)
        
        # training history
        self.loss_history_train = []
        self.acc_history_train = []
        self.loss_history_valid = []
        self.acc_history_valid = []
        self.loss_history_test = []
        self.acc_history_test = []
        
    def augment_x(self, X):
        # add extra column of 1 to X
        a_x = np.zeros((X.shape[0], X.shape[1]+1))
        a_x[:X.shape[0], :X.shape[1]] = X
        a_x[:X.shape[0], X.shape[1]] = np.ones(X.shape[0])
        return a_x
    
    def clear_history(self):
        self.loss_history_train = []
        self.acc_history_train = []
        self.loss_history_valid = []
        self.acc_history_valid = []
        self.loss_history_test = []
        self.acc_history_test = []
    
    def forward(self, X, weights=None):
        # inference 
        if X.shape[1] != self.in_dim:
            X = self.augment_x(X)
        output = np.matmul(X, self.weights) if weights is None else np.matmul(X, weights)
        return 1 / (1 + np.exp(-output)) 
    
    def regularization(self, mode="l2"):
        if mode == "l2":
            loss = np.sum(self.weights[:-1] * self.weights[:-1])
        else:     # by default l1
            loss = np.sum(np.absolute(self.weights[:-1]))
        return loss / 2.0
    
    def loss(self, y, t, minval=10**(-100)):
        # negative log-probabilities
        #logits = np.log(y.clip(min=minval))
        loss_ce = -np.sum(t * np.log(y) + (1-t) * np.log(1-y))
        loss_reg = self.regularization(mode="l2")
        return loss_ce / float(t.shape[0]) + self.reg_strength * loss_reg
    
    def classify(self, y):
        return np.asarray(y > 0.5).astype(int)
    
    def accuracy(self, y, t):
        return np.sum(np.equal(self.classify(y), t).astype(int)) / float(y.shape[0])
    
    def gradient(self, X, y, t):
        # gradients
        if X.shape[1] != self.in_dim:
            X = self.augment_x(X)
        loss_d = (y - t) / float(t.shape[0])
        w_grad_reg = self.reg_strength * self.weights
        w_grad_reg[-1] = 0
        w_grad = np.matmul(loss_d, X) + w_grad_reg
        return w_grad 
    
    def gradient_descent(self, lr=0.0001, epochs=5000, threshold=10e-8, momentum=False, gamma=0.9, Xs=None, ts=None, display=True, disp_freq=500, save_freq=500):
        # setup
        X = self.augment_x(Xs) if Xs is not None else self.X_train
        t = ts if ts is not None else self.t_train
        self.clear_history()
        v = np.zeros(self.weights.shape[0])
        lrate = lr
        #lr_decay = False
        counter = 0
        
        for i in range(epochs):
            y = self.forward(X)
            loss = self.loss(y, t)
            self.loss_history_train.append(loss)

            grad_w = self.gradient(X, y, t)
            if not momentum:
                self.weights -= lrate * grad_w
            else:
                # with momentum, discount rate gamma
                v = gamma * v + lrate * grad_w
                self.weights -= v

            acc = self.accuracy(y, t)
            self.acc_history_train.append(acc)
            
            # leanrning rate decay schedule 
            if acc > 0.95:
                if counter == 500: #and not lr_decay:
                    lrate /= 2
                    counter = 0
                else:
                    counter += 1
                #lr_decay = True
                
            # on validation set 
            loss_v, acc_v = self.evaluate(mode="valid", display=False)
            self.loss_history_valid.append(loss_v)
            self.acc_history_valid.append(acc_v)
            
            # on test set 
            loss_t, acc_t = self.evaluate(mode="test", display=False)
            self.loss_history_test.append(loss_t)
            self.acc_history_test.append(acc_t)
            
            if i % disp_freq == 0 and display:
                print("epoch %d/%d, loss %f, accuracy %f" % (i, epochs, loss, acc))
            if i % save_freq == 0:
                save_path = "models/model_e" + str(i) + ".pkl"
                self.save(path = save_path)
            if loss < threshold:
                print("loss drops below threshold %f" % threshold)
                break 
    
    def train(self, lr=0.0001, epochs=1000, threshold=10e-8, momentum=False, gamma=0.9, Xs=None, ts=None, display=True, disp_freq=500, save_freq=500, reg_strength=0.001):
        self.reg_strength = reg_strength
        self.gradient_descent(lr, epochs, threshold, momentum, gamma, Xs, ts, display, disp_freq, save_freq)
    
    def evaluate(self, mode="train", Xs=None, ts=None, display=True):
        if mode == "train":
            X, t = self.X_train, self.t_train
        elif mode == "valid":
            X, t = self.X_valid, self.t_valid
        elif mode == "test":
            X, t = self.X_test, self.t_test
        else:
            X, t = Xs, ts
        y = self.forward(X)
        loss = self.loss(y, t)
        acc = self.accuracy(y, t)
        if display:
            print("loss is %f, accuracy is %f" % (loss, acc))
        return loss, acc 
    
    def plot(self, path="images/performance.png", display=True):
        plt.figure(figsize=(20, 10))
        
        # loss curve 
        plt.subplot(1, 2, 1)
        trainl, = plt.plot(self.loss_history_train, label="train loss")
        validl, = plt.plot(self.loss_history_valid, label="valid loss")
        testl, = plt.plot(self.loss_history_test, label="test loss")
        plt.title("Loss Curve")
        plt.xlabel("epochs")
        plt.ylabel("loss")
        plt.legend([trainl, validl, testl], ["train loss", "valid loss", "test loss"]) 
                                             
        # accuracy curve 
        plt.subplot(1, 2, 2)
        traina, = plt.plot(self.acc_history_train, label="train acc")
        valida, = plt.plot(self.acc_history_valid, label="valid acc")
        testa, = plt.plot(self.acc_history_test, label="test acc")
        plt.title("Accuracy Curve")
        plt.xlabel("epochs")
        plt.ylabel("accuracy")
        plt.legend([traina, valida, testa], ["train acc", "valid acc", "test acc"])
                                             
        # display plot
        plt.grid(True)
        plt.savefig(path)
        if display:
            plt.show()
            
    def save(self, path="models/model.pkl"):
        with open(path, "wb") as f:
            pickle.dump(self.weights, f)
        print("model saved to "+path)
```

```python
def retrieve_indicators(model, num, include_bias=True, stop_words=True):
    cand_weights = model.weights
    if not include_bias:
        cand_weights = cand_weights[:-1]
    ordered_weights = sorted(list(enumerate(cand_weights)), key=lambda x: x[1], reverse=True)
    
    if not stop_words:
        ordered_weights = [(i, weight) for i, weight in ordered_weights if idx2word[i] not in ENGLISH_STOP_WORDS]
    high = ordered_weights[:num]
    low = ordered_weights[-num:][::-1]
    return (high, low)
```

H(V) = Sum(-P(V=v)logP(V=v))

H(B|A) = Sum( P(A=a) [-Sum(P(B=b|A=a) log P(B=b|A=a)] )

I(Y; X) = H(Y) - Sum(P(X=x)H(Y|X=x))

### Part 8(a)

```python
def get_feature_info(feature, datasets):
    # get feature statistics where feature is a word 
    pos_c, neg_c, pos_real_c, neg_real_c = 0, 0, 0, 0
    for w_list, label in datasets["r_train"]+datasets["f_train"]:
        if feature in w_list: 
            pos_c += 1
            if label == 1:
                pos_real_c += 1
        else:
            neg_c += 1
            if label == 1:
                neg_real_c += 1
    return pos_c, neg_c, pos_real_c, neg_real_c

def get_class_entropy(datasets):
    nums = np.asarray([len(datasets["r_train"]), len(datasets["f_train"])]) 
    probs = nums / np.sum(nums)
    return -np.sum(probs * np.log2(probs))

def get_entropy(prob_list):
    # prob list should be a list of pos_prob and neg_prob
    return -np.sum(prob_list * np.log2(prob_list))
     
def get_mutual_info(feature, y_entropy, datasets):
    # pos comes first 
    pos_c, neg_c, pos_real_c, neg_real_c = get_feature_info(feature, datasets)
    feature_nums = np.asarray([pos_c, neg_c])
    feature_probs = feature_nums / np.sum(feature_nums)
    
    pos_class_nums = np.asarray([pos_real_c, pos_c - pos_real_c])
    pos_class_probs = pos_class_nums / np.sum(pos_class_nums)
    neg_class_nums = np.asarray([neg_real_c, neg_c - neg_real_c])
    neg_class_probs = neg_class_nums / np.sum(neg_class_nums)
    
    prob_list = np.asarray([get_entropy(item) for item in [pos_class_probs, neg_class_probs]])
    cond_entropy = np.sum(feature_probs * prob_list)
    mutual_info = y_entropy - cond_entropy
    return mutual_info
```

```python
def compare_mutual_infos(datasets, idx2word, num=10):
    res = {}
#     np.random.seed(0)
    idx = np.random.choice(len(idx2word), num, replace=False)
    features = [idx2word[i] for i in idx]
    
    y_entropy = get_class_entropy(datasets)
    for f in features:
        mutual_info = get_mutual_info(f, y_entropy, datasets)
        if not np.isnan(mutual_info):
            res[f] = mutual_info
            print(f, mutual_info)
    return res 
```
